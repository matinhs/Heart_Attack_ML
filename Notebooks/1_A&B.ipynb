# Importing Packages

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Loading the Data

```python
Data = pd.read_csv('dataset_III.csv')
Data.head()
```

# Visualizing the data

```python
plt.subplot(131)
Data['output'].value_counts().plot(kind='bar',figsize = (10, 5))
plt.subplot(133)
Data['output'].value_counts().plot(kind='pie',figsize = (10, 5))
plt.suptitle('Heart attack risk, 1=High 0=low')
plt.plot()
```

```python
data1 = Data[Data['output']==1]
data2 = Data[Data['output']==0]
```

```python
plt.subplot(211)
data1['sex'].value_counts().plot(kind='pie',figsize = (10, 10))
plt.title('gender distribution in patients with high risk of heart attack')
plt.subplot(212)
data2['sex'].value_counts().plot(kind='pie',figsize = (10, 10))
plt.title('gender distribution in patients with low risk of heart attack')
plt.plot()
```

```python
plt.figure(figsize=(8, 8), dpi=80)
plt.subplot(311)
plt.hist(data1['age'], bins=15)
plt.xlabel('Age')
plt.title('age distribution in patients with high risk of heart attack')
plt.subplot(313)
plt.hist(data2['age'], bins=10)
plt.xlabel('Age')
plt.title('age distribution in patients with low risk of heart attack')
plt.plot()
```

```python
plt.figure(figsize=(8, 8), dpi=80)
plt.subplot(311)
plt.hist(data1['trtbps'], bins=10)
plt.xlabel('resting blood pressure (mm Hg)')
plt.title('resting blood pressure distribution in patients with high risk of heart attack')
plt.subplot(313)
plt.hist(data2['trtbps'], bins=10)
plt.xlabel('resting blood pressure (mm Hg)')
plt.title('resting blood pressure distribution in patients with low risk of heart attack')
plt.plot()
```

```python
plt.figure(figsize=(8, 8), dpi=80)
plt.subplot(311)
plt.hist(data1['chol'], bins=20)
plt.xlabel('cholesterol level (mg/dl)')
plt.xticks(np.arange(50, 700, 50))
plt.title('cholesterol level distribution in patients with high risk of heart attack')
plt.subplot(313)
plt.hist(data2['chol'], bins=20)
plt.xlabel('cholesterol level (mg/dl)')
plt.xticks(np.arange(50, 700, 50))
plt.title('cholesterol level distribution in patients with low risk of heart attack')
plt.plot()
```

```python
plt.subplot(211)
data1['fbs'].value_counts().plot(kind='pie',figsize = (10, 10))
plt.title('patients with high risk of heart attack')
plt.subplot(212)
data2['fbs'].value_counts().plot(kind='pie',figsize = (10, 10))
plt.title('patients with low risk of heart attack')
plt.suptitle('(fasting blood sugar > 120 mg/dl) (1 = true; 0 = false)')
plt.plot()
```

```python
plt.subplot(211)
data1['restecg'].value_counts().plot(kind='pie',figsize = (10, 10))
plt.title('patients with high risk of heart attack')
plt.subplot(212)
data2['restecg'].value_counts().plot(kind='pie',figsize = (10, 10))
plt.title('patients with low risk of heart attack')
print('resting electrocardiographic results','\n', 'Value 0: normal', '\n', 'Value 1: having ST-T wave abnormality','\n', 'Value 2: showing probable or definite left ventricular hypertrophy')
plt.plot()
```

```python
plt.figure(figsize=(10, 8), dpi=80)
plt.subplot(311)
plt.hist(data1['thalachh'], bins=15)
plt.xlabel('maximum heart rate achieved')
plt.title('maximum heart rate distribution in patients with high risk of heart attack')
plt.subplot(313)
plt.hist(data2['thalachh'], bins=15)
plt.xlabel('maximum heart rate achieved')
plt.title('maximum heart rate distribution in patients with low risk of heart attack')
plt.plot()
```
